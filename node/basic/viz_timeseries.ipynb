# 시계열 데이터 시각화

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

flights = pd.read_csv("data//flights.csv")
flights.head()
```

```python
# seaborn barplot
sns.barplot(data=flights, x="year", y="passengers", hue="year", palette="rainbow")
```

```python
# seaborn pointplot
sns.pointplot(data=flights, x="year", y="passengers")
```

```python
# seaborn line plot
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
# 달별로 나누어 보기
sns.lineplot(data=flights, x="year", y="passengers", hue="month", palette="ch:.50")
plt.legend(bbox_to_anchor=(1.03, 1), loc=2)
```

```python
# histogram
sns.histplot(data=flights["passengers"], bins=50)
```

### heatmap

```python
pivot = flights.pivot(index="year", columns="month", values="passengers")
pivot
```

```python
sns.heatmap(data=pivot)
```

```python
sns.heatmap(data=pivot, linewidths=.2)
```

```python
sns.heatmap(data=pivot, linewidths=.2, annot=True)
```

```python
sns.heatmap(data=pivot, linewidths=.2, annot=True, fmt="d")
```

```python
sns.heatmap(data=pivot, linewidths=.2, annot=True, fmt="d", cmap="YlGnBu")
```
