# 수치형 데이터 시각화

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data/tips.csv")
```

### 산점도(scatter plot)

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="total_bill", palette="ch:r=-.2,d=.3_r")
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="day")
```

### 선 그래프(line gragh)

```python
plt.plot(np.random.randn(50).cumsum()) # 표준 정규분포 누적합
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), "o")
plt.plot(x, np.cos(x))
plt.show()
```

```python
# seaborn
sns.lineplot(x=x, y=np.sin(x))
sns.lineplot(x=x, y=np.cos(x))
```

### 히스토그램

```python
# 그래프 데이터
mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)

# 축 그리기
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

patches = ax1.hist(x1, bins=50, density=False)
patches = ax1.hist(x2, bins=50, density=False, alpha=0.5)

ax1.xaxis.set_ticks_position("bottom") # x축 눈금을 아래에 표시
ax1.yaxis.set_ticks_position("left") # y축 눈금을 왼쪽에 표시

plt.xlabel("Bins")
plt.ylabel("Number of Values in Bin")
ax1.set_title("Two Frequency Distiributions")

plt.show()
```

```python
# tips 데이터 히스토그램

sns.histplot(data=df["total_bill"], label="total_bill")
sns.histplot(data=df["tip"], label="tip").legend()
```

```python
# 전체 결제 금액 대비 팁의 비율을 나타내는 히스토그램 

df["tip_pct"] = (df["tip"] / df["total_bill"]) * 100

plt.hist(df["tip_pct"], bins=50)
plt.show()
```

```python
sns.histplot(data=df, x="tip_pct", bins=100)
```

```python
# 커널 밀도 추정 그래프
df["tip_pct"].plot(kind="kde")
```
