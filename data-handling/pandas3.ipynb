### apply

```python
# apply 예시 (칼로리가 1300 보다 큰 값과 작은 값)
def cal(x):
    if x >= 1300:
        return "yes"
    else:
        return "no"
print(cal(1400))
print(cal(1200))
df["칼로리"].apply(cal)
```

```python
# apply 적용해서 새로운 컬럼 생성 (칼로리 컬럼 활용)
df["살찔까요"] = df["칼로리"].apply(cal)
df
```

```python
# lambda, apply 활용
df["고민"] = df["칼로리"].apply(lambda x: "먹지말자" if x >= 1300 else "먹자")
df
```

```python
df.to_csv("final_modudak.csv", index=False)
```

```python
# Q. 다음 df 변수의 데이터에서 각 메뉴의 이름에 '❤️'를 붙여주세요.
# ex. 황금후라이드 -> 황금후라이드❤️
# apply와 lambda를 활용해주세요.
df = pd.read_csv('new_modudak.csv')
df["메뉴"] = df["메뉴"].apply(lambda x: x+"❤️")
df
```

```python
# Q. 다음 df 변수의 데이터에서 '가성비'라는 칼럼을 만들고,
# '호수' / '가격' 의 값이 0.0008을 넘으면 '가성비' 칼럼 값을 'Good' 아니면 'Bad'로 넣어주세요.
df = pd.read_csv('new_modudak.csv')
df
ho = df["호수"].str.replace("호", "").astype(int)
cond = ho / df["가격"] > 0.0008
df["가성비"] = cond.apply(lambda x: "Good" if x else "Bad")
df
```
